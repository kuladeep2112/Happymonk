## Create a data frame that has image path and respective label


```python
path=[]
label=[]
abs_path='/content/train/'
for i in os.listdir('/content/train'):
    if int(i.split('.')[1])<4000:
        if i.split('.')[0]=='cat':
            path.append(abs_path+i)
            label.append('0')
        elif i.split('.')[0]=='dog':
            path.append(abs_path+i)
            label.append('1')
```

```python
df=pd.DataFrame(list(zip(path,label)),columns=['path','label'])
df.head()
```

## Data Generator with Augmentations

```python
datagen=ImageDataGenerator(rescale=1./255.,validation_split=0.25, 
                           rotation_range=20,
                           width_shift_range=0.2,
                           height_shift_range=0.2,
                           horizontal_flip=True,
                           shear_range=0.2,
                           zoom_range=0.1,
                           )
```

```python
train_generator=datagen.flow_from_dataframe(
    dataframe=df,
    directory="./train/",
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=True,
    seed=1,
    subset='training',
)
test_generator=datagen.flow_from_dataframe(
    dataframe=df,
    directory="./train/",
    x_col='path',
    y_col='label',
    target_size=(224, 224),
    class_mode='categorical',
    batch_size=32,
    shuffle=True,
    seed=1,
    subset='validation',
)
```

## Training of VGG net with augmentations

```python
# build the VGG16 network
input_tensor = Input(shape=(224, 224, 3))
model= applications.VGG16(include_top=False, weights='imagenet',input_tensor=input_tensor)
layer_dict = dict([(layer.name, layer) for layer in model.layers])
x = layer_dict['block5_pool'].output
conv1=Conv2D(filters=64, kernel_size=(3,3), padding='same',activation='relu')(x)
pool1=MaxPooling2D(pool_size=(2,2),padding='same')(conv1)
flatten=Flatten()(pool1)
FC1=Dense(128, activation='relu')(flatten)
FC2=Dense(64, activation='relu')(FC1)
output=Dense(2, activation='sigmoid')(FC2)
model_final = Model(input_tensor,output)
for layer in model_final.layers[:19]:
    layer.trainable = False
model_final.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_final.summary()
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size
model_final.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=test_generator,
                    validation_steps=STEP_SIZE_TEST,
                    epochs=10
)
```

## Training of Inseption net with augmentations

```python
# load model without classifier layers
model = InceptionV3(include_top=False, input_shape=(224, 224, 3))
# add new classifier layers
flat1 = Flatten()(model.layers[-1].output)
class1 = Dense(512, activation='relu')(flat1)
output = Dense(2, activation='sigmoid')(class1)
# define new model
model = Model(inputs=model.inputs, outputs=output)
for layer in model.layers[:300]:
    layer.trainable = False
# summarize
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size
model.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=test_generator,
                    validation_steps=STEP_SIZE_TEST,
                    epochs=10
)
```
